```python
from sklearnex import patch_sklearn
patch_sklearn()
import pandas as pd
import numpy as np
from pycaret.classification import *
%matplotlib inline
```

```python
sample_list_path = '../data/ABC_inter/Normalized/sample_lists/class/ABC_inter/sample_list.tsv'
sample_list = np.loadtxt(sample_list_path, delimiter='\t', dtype='<U32')
labels_path = '../data/ABC_inter/Normalized/labels.tsv'
train_df = pd.read_csv('/l/users/sayed.hashim/files/Research/Methods/contrastive-tcga/sweep/contrastive-omics/checkpoints/new_cont_data_ratio/fold-0/pretraining/version_7/latents/train_latent_space.tsv', sep='\t', header=0, index_col=0)
test_df = pd.read_csv('/l/users/sayed.hashim/files/Research/Methods/contrastive-tcga/sweep/contrastive-omics/checkpoints/new_cont_data_ratio/fold-0/pretraining/version_7/latents/test_latent_space.tsv', sep='\t', header=0, index_col=0)
```

```python
labels = pd.read_csv(labels_path, sep='\t', header=0, index_col=0)
labels = labels.loc[sample_list]
train_labels = labels.loc[train_df.index]
test_labels = labels.loc[test_df.index]
```

```python
use_outputs = True
if use_outputs:
    train_outputs = pd.read_csv('/l/users/sayed.hashim/files/Research/Methods/contrastive-tcga/sweep/contrastive-omics/checkpoints/new_cont_data_ratio/fold-0/pretraining/version_7/outputs/train_model_outputs.tsv', sep='\t', header=0, index_col=0)
    test_outputs = pd.read_csv('/l/users/sayed.hashim/files/Research/Methods/contrastive-tcga/sweep/contrastive-omics/checkpoints/new_cont_data_ratio/fold-0/pretraining/version_7/outputs/test_model_outputs.tsv', sep='\t', header=0, index_col=0)
    train_outputs.columns = range(512, 512+len(train_outputs.columns))
    test_outputs.columns = range(512, 512+len(test_outputs.columns))
    train_data = pd.concat([train_df, train_outputs, train_labels], axis=1)
    test_data = pd.concat([test_df, test_outputs, test_labels], axis=1)
else:
    train_data = pd.concat([train_df, train_labels], axis=1)
    test_data = pd.concat([test_df, test_labels], axis=1)
```

```python
exp = setup(data=train_data, target='sample_type.samples', use_gpu=True, feature_selection=False, fix_imbalance=False, normalize=False, test_data=test_data)
```
